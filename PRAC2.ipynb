```python
Q1 = df['Family_Income'].quantile(0.25)
Q3 = df['Family_Income'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['Family_Income'] >= lower) & (df['Family_Income'] <= upper)]
```

```python
plt.figure()
plt.boxplot(df['Family_Income'])
plt.title("Family Income Boxplot")
plt.show()
```

```python
Q1 = df['Assignment_Score'].quantile(0.25)
Q3 = df['Assignment_Score'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['Assignment_Score'] >= lower) & (df['Assignment_Score'] <= upper)]
```

```python
plt.figure()
plt.boxplot(df['Assignment_Score'])
plt.title('Assignment_Score')
plt.show()
```

```python
df['Log_Family_Income'] = np.log(df['Family_Income'])
```

```python
plt.figure()
plt.hist(df['Log_Family_Income'])
plt.title("Log Transformed Income")
plt.show()
```

```python
df['Normalized_Study_Hours'] = (
    df['Study_Hours_per_Week'] - df['Study_Hours_per_Week'].min()
) / (
    df['Study_Hours_per_Week'].max() - df['Study_Hours_per_Week'].min()
)
```
